```python
import pandas as pd
from sqlalchemy import create_engine, text
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
load_dotenv()
db_url= os.getenv('URL')
```

```python
db_url
```

```python
df = pd.read_excel("Online Retail Data Set.xlsx") ## Data loading
```

```python
text_columns = ["InvoiceNo", "StockCode", "Description", "Country"]
for col in text_columns:
    df[col] = df[col].astype("string")
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
engine= create_engine(db_url)
```

```python
with engine.begin() as conn:
    conn.execute(text("ANALYZE o_retail_raw"))
```

```python
df.head()  ## Data inspection
df.describe()
df.info()
```

```python
df.columns = df.columns.str.lower()
```

```python
engine= create_engine(db_url)
```

```python
df.to_sql(
    "o_retail_raw",
    engine,
    if_exists="replace",
    index=False
)
```
